```python
import os
import json
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

```python
def df_load(csv_path='C:\\Users\\amazi\\Dropbox\\ml & dl\\New folder\\train.csv'):#creating a function for converting json
    JSON_COLUMNS = ['device', 'geoNetwork', 'totals', 'trafficSource'] #these are the json columns in the given data set
    df = pd.read_csv(csv_path, 
                     converters={column: json.loads for column in JSON_COLUMNS})
    
    for column in JSON_COLUMNS: #using for loop to go in json column
        column_as_df = json_normalize(df[column]) #json normalizing is done to comvert json into seperate columns
        column_as_df.columns = [f"{column}.{subcolumn}" for subcolumn in column_as_df.columns] #it is done to naming the json columns
        df = df.drop(column, axis=1).merge(column_as_df, right_index=True, left_index=True)
    print(f"Loaded {os.path.basename(csv_path)}. Shape: {df.shape}") #optional
    return df
```

```python
%%time
train_df = df_load()
test_df = df_load("C:\\Users\\amazi\\Dropbox\\ml & dl\\New folder\\test.csv")
```
